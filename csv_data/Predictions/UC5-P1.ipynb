```python
import pandas as pd
import numpy as np
import csv
import math

personality_data = pd.read_csv('./data/PersonalityData.csv', names=["user_id", "openness", "agreeableness", "emotional_stability", "conscientiousness", "extraversion"])
rating_data = pd.read_csv('./data/Rating.csv', names=["user_id", "movie_id", "rating", "timestamp"])
rating_data = rating_data.drop("timestamp", axis=1)

print("rdf", rating_data.shape)
whole_data = pd.merge(personality_data, rating_data, left_on='user_id', right_on='user_id', how='left')
print(whole_data.head())
```

```python
def specificMovieData(table, movie_id):
    return table.loc[table['movie_id'] == movie_id]

movie_ids = sorted(set(whole_data['movie_id']))

#movie_id = [traits]
all_predicted = []
traits_to_shortened = {"openness": "open", "agreeableness": "agreeable", "emotional_stability":"emotionally_stable", "extraversion":"extraverted", "conscientiousness":"conscientious"}
missing = 0


def extrapolate(personality_rating, rating):
    if rating >= 3:
        if personality_rating >= 5:
            return True
    else:
        if 10 - personality_rating >= 5:
            return True
    return False
    

for movie_id in movie_ids:
    predicted_traits = [movie_id, "None", "None" , "None", "None" ,"None"]
    movie_data = specificMovieData(whole_data, movie_id)
    movie_data = movie_data.drop('user_id', axis=1)
    movie_data = movie_data.drop('movie_id', axis=1)
    
    movie_ratings = (movie_data.iloc[:, 5])
    
    found = False
    for i in range(0, 5):
        correlation = -1
        label = traits_to_shortened[movie_data.columns[i]]
        correlation = (movie_data.iloc[:, i]).corr(movie_ratings)
        
        if correlation > 0.7:
            predicted_traits[i + 1] = "H"
        elif correlation > 0.4:
            predicted_traits[i + 1] = "M"
        elif correlation > 0.1 :
            predicted_traits[i + 1] = "L"
        else:
            res = extrapolate(movie_data.iloc[0, i], movie_ratings.iloc[0])
            if res:
                predicted_traits[i + 1] = "L"
                
    all_predicted.append(predicted_traits)
  
'''
openness = 0
agreeableness = 1
emotional_stability = 2
conscientiousness = 3
extraversion = 4
NONE = 5
'''

    
```
